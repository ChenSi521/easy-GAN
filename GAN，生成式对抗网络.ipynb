# 什么是生成式对抗网络

生成式对抗网络是一种深度学习模型，是近年来复杂分布上无监督学习最具前景的方法之一。

**注：这里提到了无监督学习，我们的任务一般分类两种**
 - 1.监督学习，简单来说，就是数据会附带学习所需要的标签，比如MNIST集中的每张图片都带有一张标签
 - 2.非监督学习，和监督学习相反，样本不会附带学习所需的标签，需要模型自己寻找数据的分布规律

# GAN的简单原理

想要证明GAN的可收敛的话，证明会比较复杂。但是这个东西我们从理解的角度来看却又比较直观，因此这里只是简单说一下GAN的原理。

GAN模型一般由两部分组成：生成器和判别器。

从名字就可以知道，生成器就是用来生成以假乱真的数据，那么判别器又是干什么的呢？

判别器是用来判别一个样本的真假，是用来辅助我们生成器训练的。训练的过程就像博弈一样：

生成器生成出假样本，将将假样本和真样本混合在一起，投给判别器。

那么对于生成器来说，我们更希望生成器生成的所有数据都能被判别器认为是真的，而对于判别器来说，我们希望判别器能够正确的区分所有真假样本。

训练的过程，就是生成器和判别器共同进步的过程，当GAN收敛时，生成器的输出以假乱真，判别器只能通过胡猜的方式去判别，此时有判别器的准确率为50%左右。

当然，因为模型由两部分组成，所以训练可能不会如我们像的那样一帆风顺，可能会出现其中一头压着另一头打的情况，比如当生成器训练的比判别器快的话，模型收敛的速度就会降低，训练就会更慢(因为网络的收敛需要双方共同进步)

而对于判别器，我们需要将生成器生成的假数据和真数据混合在一起，让判别器判断真假，判断的过程其实就是做一个二分类，我们令输出在0~1之间，表示判别器认为数据是真实的概率。因此，我们希望假数据都生成0，真数据都生成1.
